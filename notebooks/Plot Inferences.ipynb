```python
import os
from glob import glob
import pandas as pd
import numpy as np

from PIL import Image
import matplotlib.pyplot as plt
```

```python
images = glob('../input/data/eval/images/*')
img_dict = {os.path.basename(i):i for i in images}

submission_path = "../prediction/VanillaResNet_epoch26_lr0.00025_transformbase_loss0.0037_acc0.8436_seed42.pth/submission.csv"
submission_old_path = "../prediction/vanillaresnet_epoch29_transformbase_loss0.0036_acc0.8503_seed42.pth/submission.csv"
pred = pd.read_csv(submission_path)
pred_old= pd.read_csv(submission_old_path)
```

```python
class_ = 7
pred_cat = pred[pred['ans'] == class_]['ImageID'].tolist()
if len(pred_cat) >= 20:
    pred_cat_subsample = np.random.choice(pred_cat, size=30, replace=False).tolist()
else:
    pred_cat_subsample = np.random.choice(pred_cat, size=len(pred_cat), replace=False).tolist()

for p in pred_cat_subsample:
    img = Image.open(img_dict[p])
    plt.imshow(img)
    plt.show()
```
